```python
import numpy as np
import matplotlib.pyplot as plt

def triangle_wave2012_2014(x, c0,c1,c2,c3, r0,r1,r2,r3):
    if x <= c0:
        return x *r0 *0.75;
    elif c0<x<=c1:
        return (c0* r0 +(x-c0)* r1)*0.75;
    elif c1<x<=c2:
        return (c0*r0+(c1-c0)*r1+(x-c1)*r2)*0.75;
    elif c2<x<=c3:
        return (c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)*r3)*0.75;
    else:
        return c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)* r3- 10;
    
    
def triangle_wave2014_2017(x, c0,c1,c2,c3, r0,r1,r2,r3):
    if x <= c0:
        return x *r0 *0.75;
    elif c0<x<=c1:
        return (c0* r0 +(x-c0)* r1)*0.75;
    elif c1<x<=c2:
        return (c0*r0+(c1-c0)*r1+(x-c1)*r2)*0.75;
    elif c2<x<=c3:
        return (c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)*r3)*0.75;
    else:
        return c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)*r3 - 20;

def triangle_wave2017_2018(x, c0,c1,c2,c3, r0,r1,r2,r3):
    if x <= c0:
        return x *r0 *0.75;
    elif c0<x<=c1:
        return (c0* r0 +(x-c0)* r1)*0.75;
    elif c1<x<=c2:
        return (c0*r0+(c1-c0)*r1+(x-c1)*r2)*0.75;
    elif c2<x<=c3:
        return (c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)*r3)*0.75;
    else:
        return c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)*r3 - 30;
    
    
def triangle_wave2018_2010(x, c0,c1,c2,c3,c4,r0,r1,r2,r3,r4):
    if x <= c0:
        return x *r0 *0.75;
    elif c0<x<=c1:
        return (c0* r0 +(x-c0)* r1)*0.75;
    elif c1<x<=c2:
        return (c0*r0+(c1-c0)*r1+(x-c1)*r2)*0.75;
    elif c2<x<=c3:
        return (c0*r0+(c1-c0)*r1+(c2-c1)*r2+(x-c2)*r3)*0.75;
    elif c3<x<=c4:
        return (c0*r0+(c1-c0)*r1+(c2-c1)*r2+(c3-c2)*r3+(x-c3)*r4)*0.75;
    else:
        return c0*r0+(c1-c0)*r1+(c2-c1)*r2+(c3-c2)*r3+(x-c3)*r4-20.000;

    
x = np.linspace(0, 1500, 2000)   
    
y1 = np.array([triangle_wave2012_2014(t, 40.000, 80.000, 120.000,305.882353, 0.02, 0.07, 0.12, 0.17) for t in x])
y2 = np.array([triangle_wave2014_2017(t, 40.000, 80.000, 120.000,541.176471, 0.02, 0.07, 0.12, 0.17) for t in x])
y3 = np.array([triangle_wave2017_2018(t, 45.000, 90.000, 135.000,785.294118, 0.02, 0.07, 0.12, 0.17) for t in x])
y4 = np.array([triangle_wave2018_2010(t, 50.000, 100.000, 150.000, 200.000, 576.470588,0.02, 0.06, 0.1, 0.14, 0.17) for t in x])
# y5 = np.array([triangle_wave2019_2020(t, 50.000, 100.000, 150.000, 200.000, 576.470588, 0.02, 0.06, 0.1, 0.14, 0.17) for t in x])
    
    
# plt.style.use('dark_background')
plt.rcParams.update({'font.size': 16})
plt.figure()
plt.plot(x, y1,'xkcd:golden rod',label='2012-2014')
plt.plot(x, y2,'xkcd:lavender blue',label='2014-2017')
plt.plot(x, y3,'xkcd:shamrock green',label='2017-2018')
plt.plot(x, y4,'tab:red',label='2018-2020')

plt.ylim(0, 200)  
fig_size = plt.rcParams["figure.figsize"]
print("Current size:", fig_size)
fig_size[0] = 16.0
fig_size[1] = 8.0
plt.rcParams["figure.figsize"] = fig_size
print("Current size:", fig_size)
plt.xlabel("\nAnnual Income (HKD in thousand)")
plt.ylabel("\nIncome Tax to pay (HKD in thousand)")
plt.title("\nIncome Tax Payable Amount\n over past 7 years\n")
plt.legend()
plt.show()
# plt.savefig('tax_light.png') 
print("Done!")
plt.close()
```
